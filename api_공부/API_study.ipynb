- API(애플리케이션 프로그래밍 인터페이스, Application Programming Interface)는 소프트웨어 응용 프로그램들이 서로 상호작용할 수 있도록 하는 정의 및 프로토콜의 집합

- 업비트에서 제공하는 API 문서를 참고하여 적절한 엔드포인트를 지정해야 한다.
- https://docs.upbit.com/reference/%EB%A7%88%EC%BC%93-%EC%BD%94%EB%93%9C-%EC%A1%B0%ED%9A%8C

```python
import requests
headers = {"accept": "application/json"}
```

```python
url = "https://api.upbit.com/v1/market/all?isDetails=false"
response = requests.get(url, headers=headers)

for crypto in eval(response.text): # crypto : 암호화폐
   print(crypto)
```

```python
# 다른 방법
import json

# JSON 데이터를 안전하게 파싱
data = json.loads(response.text)

for crypto in data:
    print(crypto)
```

```python
import json

url = "https://api.upbit.com/v1/market/all?isDetails=true" # 유의종목 필드과 같은 상세 정보 노출 여부를 true로 지정
response = requests.get(url, headers=headers)

# JSON 데이터를 안전하게 파싱
data = json.loads(response.text)

markets = [] # 현재가 조회하려는 market 리스트

for crypto in data:
    if crypto['market_warning'] == 'NONE' and crypto['market'].startswith('KRW'): # 원화시장의 투자유의가 아닌 종목만 받아오기
        print(crypto)
        markets.append(crypto['market']) 

markets = ", ".join(markets)
```

```python
url = 'https://api.upbit.com/v1/ticker?markets=KRW-BTC' # 현재가 조회, 마켓코드는 반점으로 구분됨
response = requests.get(url, headers=headers)

ticker = eval(response.text)[0]

print(ticker['trade_price']) # 종가
```

```python
url = f'https://api.upbit.com/v1/ticker?markets={markets}' # 현재가 조회, 마켓코드는 반점으로 구분됨
response = requests.get(url, headers=headers)

for ticker in eval(response.text):
    print(f"{ticker['market']} : {ticker['trade_price']}")
```

```python
url = 'https://api.upbit.com/v1/candles/months?market=KRW-BTC&count=3' # 월 캔들 3개 조회
response = requests.get(url, headers=headers)

for result in eval(response.text):
    print(f"date : {result['first_day_of_period']}")
    print(f"open : {result['opening_price']}")
    print(f"high : {result['high_price']}")
    print(f"low : {result['low_price']}")
    print(f"close : {result['trade_price']}")
    print(f"volume : {result['candle_acc_trade_volume']}")
    print()
```

```python
url = 'https://api.upbit.com/v1/candles/days?market=KRW-BTC&count=3' # 일 캔들 3개 조회
response = requests.get(url, headers=headers)

for result in eval(response.text):
    print(f"date : {result['candle_date_time_kst']}")
    print(f"open : {result['opening_price']}")
    print(f"high : {result['high_price']}")
    print(f"low : {result['low_price']}")
    print(f"close : {result['trade_price']}")
    print(f"volume : {result['candle_acc_trade_volume']}")
    print()
```

```python
url = "https://api.upbit.com/v1/orderbook?markets=KRW-BTC"   # 호가 보기
response = requests.get(url, headers=headers)

orderbook_units = eval(response.text)[0]['orderbook_units']

for order in orderbook_units:
    print(order)
```
