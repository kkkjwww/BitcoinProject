```python
import pandas as pd
import numpy as np

import ta
import ta.utils
import ta.volatility as tav

import requests

import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
```

## 전략 만들기

```python
# 캔들 200개씩 받아서 1년간 일봉 데이터 받는 함수
def get_1day_sticks_of_1year(coin, year):
    url = f"https://api.upbit.com/v1/candles/days?market=KRW-{coin}&to={year}-12-31%2012%3A59%3A59&count=200"
    headers = {"accept": "application/json"}
    response = requests.get(url, headers=headers)
    print(response)
    df = pd.DataFrame(response.json()).sort_values('candle_date_time_utc').reset_index().drop(columns='index')

    df2 = pd.DataFrame()
    url = f"https://api.upbit.com/v1/candles/days?market=KRW-{coin}&to={year}-06-14%2012%3A59%3A59&count=200"
    headers = {"accept": "application/json"}
    response = requests.get(url, headers=headers)
    print(response)
    df2 = pd.DataFrame(response.json()).sort_values('candle_date_time_utc').reset_index().drop(columns='index')   
    
    result = pd.concat([df2,df]).reset_index().drop(columns='index').drop_duplicates()
    return result
```

```python
def get_profit_rate(df):

    # DMI, ADX 지표
    adx = ta.trend.ADXIndicator(high=df['high_price'], low=df['low_price'], close=df['trade_price'])
    
    df['ADX'] = adx.adx()
    df['PDI'] = adx.adx_pos()
    df['MDI'] = adx.adx_neg()
    df['candle_date_time_utc'] = pd.to_datetime(df['candle_date_time_utc'])

    df['SMA_10'] = df['ADX'].rolling(window=10).mean()  # ADX 이평선
    df['ADX_SMA_width'] = df['ADX'] - df['SMA_10'] # ADX 이평선과의 차이(너비)
    df.loc[35, 'ADX_SMA_width'] = df.loc[36, 'ADX_SMA_width'] # 결측값 채우기
    df['ADX_SMA_slope'] = (df['ADX_SMA_width'] - df['ADX_SMA_width'].shift()) / 2 # 너비 평균 변화율
    df.loc[35, 'ADX_SMA_slope'] = df.loc[36, 'ADX_SMA_slope'] # 결측값 채우기


    # 볼린져 지표
    df['upper'] = tav.bollinger_hband(df['trade_price'])
    df['lower'] = tav.bollinger_lband(df['trade_price'])
    df['middle'] = tav.bollinger_mavg(df['trade_price'])
    df['bol_width'] = tav.bollinger_wband(df['trade_price'])
    df = df[27:].reset_index(drop=True) # 결측값 채우기

    df['bol_width_slope'] = (df['bol_width'] - df['bol_width'].shift()) / 2 # 밴드폭 평균 변화율

    
    #매수타이밍
    df['ADX_sign'] = False # ADX 기준 매수 타이밍
    df.loc[df['ADX_SMA_slope'] >= 0, 'ADX_sign'] = True  # ADX, ADX 이평선 간 폭이 커질 때

    df['DMI_sign'] = False
    df.loc[df['PDI'] - df['MDI'] > 0, 'DMI_sign'] = True # PDI가 MDI보다 클 때

    df['bol_shrink'] = False
    df['break_point'] = False
    df.loc[df['bol_width_slope'] < 0, 'bol_shrink'] = True # 밴드폭 수축할 때
    df.loc[df['trade_price'] >= df['upper'], 'break_point'] = True # 종가가 밴드 상한보다 높을 때

    breakpoint_found = False
    
    for i in range(len(df)):
        if df.loc[i, 'break_point']: # 종가가 밴드 상한선 보다 높을 때
            if not breakpoint_found: # 그 때, breakpoint_found 변수가 False 일 때
                breakpoint_found = True            
            else:
                df.loc[i, 'break_point'] = False # breakpoint_found 변수가 True이면 breakpoint 찾은 상태이므로 그 사이 break_point False로 만들어준다.
                continue
        else:  # break_point가 아닐 때
            if df.loc[i, 'bol_shrink']: # 그 때, 밴드 폭이 수축한다면 
                breakpoint_found = False # breakpoint_found False로 변경. 밴드 폭 수축 후 breakpoint 발생하므로
            else:
                continue
    df.loc[df['break_point'] & df['ADX_sign'] & df['DMI_sign'], 'stat']= 'buy'

    
    # 매도타이밍
    df['comparison_ADX'] = df.apply(lambda row: '+' if row['ADX'] - row['SMA_10'] >= 0 else '-', axis=1)
    df.loc[(df['comparison_ADX'].shift(1) == '+') & (df['comparison_ADX'] == '-'), 'stat'] = 'sell'  # ADX가 ADX 이평선 하향 돌파시 매도
    

    # 트레이딩 결과
    df_trade_info = pd.concat([df.loc[df['stat']=='buy'], df.loc[df['stat']=='sell']]).sort_index().reset_index(drop=True)
    
    trade_fee = 0.0005 # 거래 수수료 0.05%
    stat = 'buy'
    trade_lst = []
    for i in range(len(df_trade_info)):    
        if stat == 'buy':  # 매수할 차례일 떄
            if df_trade_info.loc[i ,'stat'] == 'buy': # 매수 타이밍이라면 매수 시점과 가격을 사전에 저장
                stat = 'sell'
                trade_dict = {}
                trade_dict['buy_date'] = df_trade_info.loc[i ,'candle_date_time_utc']
                trade_dict['buy'] = df_trade_info.loc[i ,'trade_price'] * (1 + trade_fee)  # 수수료만큼 더해준다.
            
            else:  # 매수할 차례인데 매수 타이밍이 아니라면 trade_info에서 지운다.
                df_trade_info.drop(index=i, inplace=True)

        else:  # 메도할 차례일 때
            if df_trade_info.loc[i ,'stat'] == 'sell': # 매도 타이밍이라면 매도 시점과 가격을 사전에 저장
                stat = 'buy'
                trade_dict['sell_date'] = df_trade_info.loc[i ,'candle_date_time_utc']
                trade_dict['sell'] = df_trade_info.loc[i ,'trade_price'] * (1 - trade_fee)  # 수수료만큼 빼준다.

                trade_lst.append(trade_dict)
                
            else:
                df_trade_info.drop(index=i, inplace=True)

    df_trade_result = pd.DataFrame(trade_lst)
    df_trade_result['profit'] = df_trade_result['sell'] - df_trade_result['buy']
    df_trade_result['profit_rate'] = df_trade_result['profit'] / df_trade_result['buy']

    return df_trade_result
```

```python
# 코인 리스트 가져오기

url = "https://api.upbit.com/v1/market/all?isDetails=true"

headers = {"accept": "application/json"}

res = requests.get(url, headers=headers)

df = pd.DataFrame(res.json())
df = df.loc[df['market'].str.contains('KRW')]
df['market'] = df['market'].str.strip('KRW-')
coinlist = list(df['market'])

# 2018년부터 2023년까지 모든 코인 전략 사용할 시 매수, 매도 포인트만 따로 뽑아서 데이터프레임 저장
years = list(range(2018, 2024))

for year in years:
    df_total_trade = pd.DataFrame() 
    fail_list = []   
    for coin in coinlist:
        try:
            df = get_1day_sticks_of_1year(coin, year)
            df_trade = get_profit_rate(df)
            df_trade.insert(0, 'coin', coin)
            df_total_trade = pd.concat([df_total_trade, df_trade])
            print(coin, '매매 정보 입력완료')
        except:
            fail_list.append(coin)
            continue
    # df_total_trade.to_csv(f'종합투자전략_매매정보/{year}_total_trade_DATA.csv', index=False)
```

## 전략 평가

```python
temp = df_total_trade.copy()
temp = temp.reset_index(drop=True)
```

```python
# 비트코인 2018년부터 2023년까지 받아오기
years = ['2018', '2019', '2020', '2021', '2022', '2023']
upbits = []

for year in years:
    df = pd.read_csv(f'DATA/upbit_{year}_1d_krw_BTC.csv')
    upbits.append(df)

upbit = pd.concat(upbits, ignore_index=True).drop_duplicates(ignore_index=True)

upbit = upbit[upbit['candle_date_time_utc'] >= '2018-01-01']
upbit = upbit.reset_index(drop=True)
upbit['candle_date_time_utc'] = pd.to_datetime(upbit['candle_date_time_utc'])
```

```python
buy_date_BTC = temp.loc[temp['coin'] == 'BTC', 'buy_date'].values
buy_price_BTC = temp.loc[temp['coin'] == 'BTC', 'buy'].values

sell_date_BTC = temp.loc[temp['coin'] == 'BTC', 'sell_date'].values
sell_price_BTC = temp.loc[temp['coin'] == 'BTC', 'sell'].values

plt.figure(figsize=(20, 10))

plt.plot(upbit['candle_date_time_utc'], upbit['trade_price'], label='BTC_price')

plt.scatter(buy_date_BTC, buy_price_BTC, label='buy', color='r')
plt.scatter(sell_date_BTC, sell_price_BTC, label='sell', color='b')

plt.ticklabel_format(axis='y',useOffset=False, style='plain')
plt.title("비트코인 가격 흐름과 혼합 전략 사용시 매수, 매도 포인트", fontsize=25)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.legend(fontsize=20)
plt.show()
```

```python
# 각 종목 누적 수익률
res_prod = temp.groupby('coin')['profit_rate'].apply(lambda x: ((x+1).prod() - 1) * 100)

# 각 종목 평균 수익률
res_mean = temp.groupby('coin')['profit_rate'].mean() * 100
```

```python
temp['buy_date'] = pd.to_datetime(temp['buy_date'])

# 년도 열 생성
temp['year'] = temp['buy_date'].dt.year

# 각 년도 코인별 누적 수익률
res_year_prod = temp.groupby(['year', 'coin'])['profit_rate'].apply(lambda x: ((x+1).prod() - 1) * 100)

# 각 년도 별로 모든 코인에 1원씩 투자했을 때 최종 잔고
res_year_prod = res_year_prod.to_frame().groupby('year')['profit_rate'].apply(lambda x: (1 + 1 * (x/100)).sum()).to_frame()\
    .rename(columns={'profit_rate': 'balance'})
# 각 년도별 투자한 코인 종목 수 (총 몇원 투자했는지를 나타냄)
how_many = temp.groupby('year')['coin'].apply(lambda x: len(x.unique())).to_frame()
res_year_prod = res_year_prod.merge(how_many, left_index=True, right_index=True, how='left')
res_year_prod['result'] = res_year_prod['balance'] - res_year_prod['coin']
res_year_prod
```

```python
# coin + coin * x = balance
# 각 년도 수익률 계산
print(f"총 번 돈 : {res_year_prod['result'].sum()} 원")
res_rate = ((res_year_prod['balance'] - res_year_prod['coin']) / res_year_prod['coin']) * 100
res_rate
```

```python
fig, ax1 = plt.subplots()

# 첫 번째 축: 막대 그래프
bars1 = ax1.bar(res_year_prod.index, res_year_prod['balance'], label='최종 잔고', color='cyan')
bars2 = ax1.bar(res_year_prod.index, res_year_prod['coin'], alpha=0.5, label='투자금액', color='green')
ax1.set_ylabel('원', rotation=0)

# 두 번째 축: 선 그래프
ax2 = ax1.twinx()
line = ax2.plot(res_rate.index, res_rate, alpha=0.7, label='수익률', color='r', marker='o')
ax2.set_ylabel('%', color='r', rotation=0)
ax2.tick_params('y', colors='r')
ax2.set_yticks(np.arange(-50, 400, 50))

# 두 축의 핸들과 레이블을 결합하여 하나의 범례 생성
handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
handles = handles1 + handles2
labels = labels1 + labels2

ax1.legend(handles, labels, loc='upper right')

ax1.set_xlabel("년도")
plt.title("각 년도별 종목별 1원씩 투자하였을 때 잔고와 수익률")
plt.show()
```

```python
# 코인별로 초기에 1원 투자했을 때 잔고 변화를 2023년까지 데이터 프레임에 저장
temp['cumprod'] = temp.groupby('coin', group_keys=False)['profit_rate'].apply(lambda x: (x+1).cumprod())
```
