## 5분봉 받아오기 예시 (2024-07-01)
- 5분봉으로 하루에 144개 캔들 존재

```python
url = "https://api.upbit.com/v1/candles/minutes/5?market=KRW-BTC&to=2024-07-01T12%3A00%3A00&count=144"
headers = {"accept": "application/json"}
response = requests.get(url, headers=headers)

df = pd.DataFrame(response.json()).sort_values('candle_date_time_utc').reset_index().drop(columns='index')
df
```

## 5분봉 받아와서 csv 파일에 저장하기 (2022년, 2023년)

```python
# 특정 날짜 입력받아 그 날의 5분봉 데이터 데이터 프레임으로 반환하는 함수
def get_5min_stick_of_1day(date):
    dfs = []
    for _ in range(2):
        url = f"https://api.upbit.com/v1/candles/minutes/5?market=KRW-BTC&to={date}T12%3A00%3A00&count=144"
        headers = {"accept": "application/json"}
        response = requests.get(url, headers=headers)
        print(response)
        df = pd.DataFrame(response.json()).sort_values('candle_date_time_utc').reset_index().drop(columns='index')
        dfs.append(df)

    result = pd.concat(dfs).reset_index().drop(columns='index')
    return result
```

```python
# 날짜 범위 입력해서 리스트 반환하는 함수
def date_range(start, end):
    dates = pd.date_range(start=start, end=end)
    return dates.strftime("%Y-%m-%d").tolist()

dates = date_range("2024-06-01", "2024-07-01")
dates
```

```python
# 날짜 리스트 입력받아 그 기간 5분봉 데이터 반환하는 함수
def get_5min_stick_of_interval(dates):
    df = pd.DataFrame()
    for date in dates:
        df2 = get_5min_stick_of_1day(date)
        df = pd.concat([df,df2])
        print(date)
    return df
    # return df.reset_index().drop(columns='index')
```

```python
# 일주일 6초
# 2023년 5분봉 데이터 받아오기
dates = date_range("2022-01-01", "2022-12-31")
df = get_5min_stick_of_interval(dates)
df
```
