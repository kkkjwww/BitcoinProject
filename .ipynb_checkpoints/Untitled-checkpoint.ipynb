```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

import requests

import os

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# 코인 리스트 받아오기
url = 'https://api.upbit.com/v1/market/all?isDetails=true'

headers = {"accept": "application/json"}
res = requests.get(url, headers=headers)

df = pd.DataFrame(res.json())
df = df.loc[df['market'].str.contains('KRW')]
df['market'] = df['market'].str.strip('KRW-')
coinlist = list(df['market'])
coinlist
```

```python
# 코인 별 파일 받아와서 하나의 데이터 프레임으로 합치기
dfs = []

for coin in coinlist:
    file_name = f'coin_1day/upbit_{coin}_prices_2023_days.csv'
    
    if os.path.exists(file_name):
        df = pd.read_csv(file_name).drop('Unnamed: 0', axis=1)
    else:
        continue
    
    # 빈 파일이 있다.
    if len(df) != 0:
        condition = (df['candle_date_time_utc'] >= '2022-12-31') & (df['candle_date_time_utc'] <= '2023-12-31')
        df = df[condition]
    dfs.append(df)
    
df_coins = pd.concat(dfs, ignore_index=True)
df_coins.head()
```

```python
df_coins = df_coins[df_coins['candle_date_time_utc'] >= '2023-01-01']
df_coins.head()
```

```python
df_coins = df_coins.pivot(index='market', columns='candle_date_time_utc', values='change_rate')
df_coins.head()
```

```python
df_coins = df_coins.dropna()
```

```python
df_coins.shape
```

```python
X = df_coins.values

kmeans = KMeans(n_clusters=88, random_state=42)
kmeans.fit(X)

labels = kmeans.predict(X)

centroids = kmeans.cluster_centers_

silhouette_avg = silhouette_score(X, labels)
print(f"실루엣 계수 : {silhouette_avg: .4f}")
```

```python
plt.scatter(X[:, 2], X[:, 5], c=labels, cmap='viridis', marker='o')
```
